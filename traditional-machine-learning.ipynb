```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


import torch
from torch import optim
from torch import nn
from torch.utils.data import TensorDataset, DataLoader, random_split
from tqdm import tqdm
import torch.optim.lr_scheduler as lr_scheduler

# !pip install torchvision
import torchvision

import torch.nn.functional as F
import torchvision.datasets as datasets
import torchvision.transforms as transforms

# !pip install torchmetrics
import torchmetrics

from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
images = torch.load("/kaggle/input/moai-2025-training/train_images.pt", weights_only = True)
labels = pd.read_csv("/kaggle/input/moai-2025-training/train_labels.csv")

images = images.float()
images = (images - images.min()) / images.max()
images = images.reshape(images.shape[0], -1) #-1是自動計特徵數, 變成二維是因為logistic regression要
images = images.numpy() #logistic regression要numpy

labels = labels.iloc[:, 1]
model = LogisticRegression(max_iter=1000, multi_class='multinomial', solver="lbfgs")
model.fit(images, labels)
print(model.score(images, labels))
```
